``` python

```
If any of these are unspecified, they default to the values ``start=0``, ``stop=``*``size of dimension``*, ``step=1``.
We'll take a look at accessing sub-arrays in one dimension and in multiple dimensions.

All of the preceding routines worked on single arrays. It's also possible to combine multiple arrays into one, and to conversely split a single array into multiple arrays. We'll take a look at those operations here.

Concatenation, or joining of two arrays in NumPy, is primarily accomplished using the routines ``np.concatenate``, ``np.vstack``, and ``np.hstack``.
``np.concatenate`` takes a tuple or list of arrays as its first argument, as we can see here:

## Introducing UFuncs

For many types of operations, NumPy provides a convenient interface into just this kind of statically typed, compiled routine. This is known as a *vectorized* operation.
This can be accomplished by simply performing an operation on the array, which will then be applied to each element.
This vectorized approach is designed to push the loop into the compiled layer that underlies NumPy, leading to much faster execution.

````python
#Consider following loop based implementation to ind reciprocals for each element of an array
np.random.seed(0)

def compute_reciprocals(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0 / values[i]
    return output

big_array = np.random.randint(1, 100, size=1000000)
%timeit compute_reciprocals(big_array)
````

````python
#Same operatoin using UFuncs applying '/' over array elements
%timeit (1.0 / big_array)
````

Vectorized operations in NumPy are implemented via *ufuncs*, whose main purpose is to quickly execute repeated operations on values in NumPy arrays. UFuncs are extremely flexible – before we saw an operation between a scalar and an array, but we can also operate between two arrays as well as multidimentional arrays.

Computations using vectorization through ufuncs are nearly always more efficient than their counterpart implemented using Python loops, especially as the arrays grow in size.
Any time you see such a loop in a Python script, you should consider whether it can be replaced with a vectorized expression.

## Answers to Questions

### 1. What are UFuncs in NumPy?

**UFuncs** (Universal Functions) in NumPy are functions that operate element-by-element on `ndarray` objects. They provide a fast, vectorized way to perform mathematical operations on entire arrays without writing explicit Python loops. This push of the loop into the compiled layer that underlies NumPy leads to much faster execution, especially for large arrays.

UFuncs can be unary (operating on a single input, like `np.abs()`) or binary (operating on two inputs, like `np.add()` or the `+` operator). Examples include arithmetic operators (`+`, `-`, `*`, `/`), trigonometric functions (`np.sin`, `np.cos`), exponential and logarithmic functions, and more specialized functions found in `scipy.special`.

### 2. Which are various attributes of NumPy array objects?

NumPy array objects (`ndarray`) have several important attributes that provide information about their structure and data type. Some key attributes are:

*   **`ndim`**: The number of dimensions (axes) of the array.
*   **`shape`**: A tuple indicating the size of each dimension.
*   **`size`**: The total number of elements in the array.
*   **`dtype`**: The data type of the elements in the array (e.g., `int32`, `float64`).
*   **`itemsize`**: The size in bytes of each element of the array.
*   **`nbytes`**: The total bytes consumed by the elements of the array.

### 4. Consider an array object named `Obj` with dimensions 3×4×2 and you want to convert it to 4×6 shape. Explain how you will do it. Show this using a code cell, taking an example of a 3×4×2 NumPy array of random integers between 1 and 50.

To convert an array from one shape to another, you can use the `reshape()` method. The crucial condition is that the total number of elements in the array must remain the same before and after reshaping. For a 3x4x2 array, the total elements are `3 * 4 * 2 = 24`. For a 4x6 shape, the total elements are `4 * 6 = 24`. Since the total number of elements matches, we can reshape the array.

First, we'll create the 3x4x2 array using `np.random.randint` and then apply the `reshape` method.

````python
import numpy as np

# Create a 3x4x2 array of random integers between 1 and 50
Obj = np.random.randint(1, 51, size=(3, 4, 2))
print("Original Obj (3x4x2):")
print(Obj)
print("Original shape:", Obj.shape)

# Reshape Obj to 4x6
Obj_reshaped = Obj.reshape(4, 6)
print("\nObj_reshaped (4x6):")
print(Obj_reshaped)
print("Reshaped shape:", Obj_reshaped.shape)
````

### 5. Consider the above 4×6 array, sample a 2×3 sub-array from the left bottom of this array and store the result in a variable named `subObjCpy`.

To sample a 2x3 sub-array from the left bottom of the 4x6 array, we need to select the last two rows and the first three columns. The 4x6 array has rows indexed 0, 1, 2, 3 and columns indexed 0, 1, 2, 3, 4, 5.

*   Last two rows: indices `2:` (from index 2 to the end)
*   First three columns: indices `:3` (from index 0 up to, but not including, index 3)

So, the slice will be `Obj_reshaped[2:, :3]`.

````python
# Sample a 2x3 sub-array from the left bottom of Obj_reshaped
subObjCpy = Obj_reshaped[2:, :3]
print("subObjCpy (2x3 from left bottom):")
print(subObjCpy)
````

### 6. Make changes in one particular value in `subObjCpy`. Check values of `Obj` and `subObjCpy` to check the impact of this change on the original and copied object. What do you conclude?

When you create a sub-array using slicing in NumPy, you are typically creating a *view* into the original array's data, not a separate copy. This means that if you modify an element in `subObjCpy`, the corresponding element in the original `Obj_reshaped` (and thus `Obj`) will also be modified, and vice-versa.

Let's change a value in `subObjCpy` and observe the effect on `Obj_reshaped`.

````python
# Change a value in subObjCpy
# For example, change the element at row 0, column 0 of subObjCpy
# This corresponds to row 2, column 0 of Obj_reshaped
subObjCpy[0, 0] = 999

print("\nsubObjCpy after modification:")
print(subObjCpy)

print("\nObj_reshaped after modification to subObjCpy:")
print(Obj_reshaped)

print("\nSpecifically, checking Obj_reshaped[2, 0]:", Obj_reshaped[2, 0])
````

**Conclusion:** As you can see, changing `subObjCpy[0, 0]` to `999` also changed the value at `Obj_reshaped[2, 0]`. This demonstrates that `subObjCpy` is a **view** of `Obj_reshaped`, not an independent copy. Changes made to the view are reflected in the original array.

### 7. If you were to filter a sub-array view from the above `Obj` object and want to apply modification in the original and/or sub-object and change the respective element in the other, how will you achieve it?

If you want to create a sub-array that is an *independent copy* of the original array's data, so that changes to the sub-array do **not** affect the original (and vice-versa), you must explicitly use the `.copy()` method when creating the sub-array. This ensures that a new memory block is allocated for the sub-array's data.

Let's create a copy of the same sub-array and demonstrate the effect of modification.

````python
# Reset Obj_reshaped to its state before the previous modification for clarity
Obj_reshaped_original = Obj.reshape(4, 6)
print("Original Obj_reshaped before new sub-array creation:")
print(Obj_reshaped_original)

# Create a sub-array using .copy()
subObjTrulyCpy = Obj_reshaped_original[2:, :3].copy()
print("\nsubObjTrulyCpy (an explicit copy):")
print(subObjTrulyCpy)

# Modify a value in subObjTrulyCpy
subObjTrulyCpy[0, 0] = 777

print("\nsubObjTrulyCpy after modification:")
print(subObjTrulyCpy)

print("\nObj_reshaped_original after modification to subObjTrulyCpy:")
print(Obj_reshaped_original)

print("\nSpecifically, checking Obj_reshaped_original[2, 0]:", Obj_reshaped_original[2, 0])
````

**Conclusion:** This time, when `subObjTrulyCpy[0, 0]` was changed to `777`, the value in `Obj_reshaped_original[2, 0]` remained unchanged. This confirms that `subObjTrulyCpy` is an independent copy, and modifications to it do not propagate back to the original array. If you desire this behavior, always use `.copy()` after slicing.
